**CUMULATIVE DISTRIBUTION FUNCTION (CDF)**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
Habbs = pd.read_csv("habberman.csv")
Habbs.columns =["age", "operation_year", "nodes", "survival_status"] 
```

```python
counts,bin_edges=np.histogram(Habbs['age'],bins = 10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print (bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("PDF and CDF plot of patients age ",color ='green',size  = 15 )
plt.ylabel('''Probability of RV X <= x for CDF
Probability of RV X = x for PDF''')
plt.xlabel("Patients age")#correction (labels added)
Label = ["pdf","cdf"]
plt.legend(Label) #correction (Legend added)
plt.show();

counts,bin_edges=np.histogram(Habbs['operation_year'],bins = 10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print (bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("PDF and CDF plot of patients operation_year  ",color ='green',size  = 15 )
plt.ylabel('''Probability of RV X <= x for CDF
Probability of RV X = x for PDF''')
plt.xlabel("Patients operation_year") #correction (labels added)
Label = ["pdf","cdf"]
plt.legend(Label) #correction (Legend added)
plt.show();

counts,bin_edges=np.histogram(Habbs['nodes'],bins = 10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print (bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("PDF and CDF plot of patients nodes ",color ='green',size  = 15 )
plt.ylabel('''Probability of RV X <= x for CDF
Probability of RV X = x for PDF''')
plt.xlabel("patients nodes") #correction (labels added)
Label = ["pdf","cdf"]
plt.legend(Label) #correction (Legend added)
plt.show();

print("Observation: 80 % of time patients survived having less than 10 nodes.")
```
